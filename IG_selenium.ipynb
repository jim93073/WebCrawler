```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import datetime

class IgSelenium():
    def __init__(self):    
        
        self.username = ''
        self.password = ''
        self.tags = "#love"    
        
        #Initial webdriver
        profile = webdriver.FirefoxProfile()
        profile.set_preference("dom.webnotifications.enabled", False)     
        profile.update_preferences()     
        self.driver = webdriver.Firefox(firefox_profile=profile)   
        
            
    
    def get_htmltext(self):
        self.driver.get("https://www.instagram.com/?hl=zh-tw")
        time.sleep(3)
        self.driver.find_element_by_name("username").send_keys(self.username)
        self.driver.find_element_by_name("password").send_keys(self.password)
        self.driver.find_element_by_xpath("//button[@type='submit']").click()
        time.sleep(3)    
        self.driver.find_element_by_xpath("//button[@type='button']").click() #click å„²å­˜è³‡æ–™    
        time.sleep(3)

        #Search
        self.driver.find_element_by_xpath("//input[@type='text'][@placeholder='æœå°‹']").send_keys(self.tags)
        time.sleep(3)
        self.driver.find_element_by_xpath("//input[@type='text'][@value='"+self.tags+"']").send_keys(Keys.ENTER,Keys.ENTER)
        time.sleep(3)

        #Get first artical index
        htmltext = self.driver.page_source   
        soup = BeautifulSoup(htmltext, 'html.parser')
        tabindex = soup.find_all('a',tabindex='0')

        index = 0
        for data in tabindex:
            if(len(data.text) == 0): # tabindex==0  is first artical
                break;
            index+=1

        print("First artical index:",index)    

        
        href = tabindex[index].get('href')
        self.driver.find_element_by_xpath ("//a[@href='" +href+ "']/div").click() #Click first artical
        
        #getConent
#         self.getContent(10)

        #Click Good 
#         self.clickGood(10)

        

    def clickGood(self,number):
        for i in range(number):
            time.sleep(1)
            try: #Page may loading failed,direct to go to next page
                self.driver.find_element_by_xpath ('//span/*[name()="svg"][@aria-label="è®š"]/parent::*/parent::*/parent::button').click()    
            except:
                print("Loading Failed")
            self.driver.find_element_by_link_text("ä¸‹ä¸€å€‹").click()
            
    def getContent(self,number):
        for i in range(number):
            time.sleep(1)
            #Get comment
            htmltext = self.driver.page_source   
            soup = BeautifulSoup(htmltext, 'html.parser')
            h2 = soup.find_all('h2')
            h2_className = h2[-1].get('class')[0]        
            
            try: #IG classåç¨±æœ€å¾Œå¯èƒ½æœ‰ç©ºç™½éµ ä½†getä¸åˆ°        
                print(self.driver.find_element_by_xpath ("//*[@class='" +h2_className+ " ']/following-sibling::span[1]").text) #è‡ªå·±åœ¨æœ€å¾ŒåŠ ç©ºç™½éµ 
            except:
                pass
            try:
                print(self.driver.find_element_by_xpath ("//*[@class='" +h2_className+ "']/following-sibling::span[1]").text) #è‡ªå·±åœ¨æœ€å¾ŒåŠ ç©ºç™½éµ 
            except:
                pass
            self.driver.find_element_by_link_text("ä¸‹ä¸€å€‹").click()
IgSelenium().get_htmltext()
```
